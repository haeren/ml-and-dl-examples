```python
from pathlib import Path
from PIL import Image

imagePath = Path('/content/drive/MyDrive/YuksekLisans/DLAPP/HW2/cat.jpg')
image = Image.open(imagePath) # Load the original image
display(image)
```

```python
flippedVertically = image.transpose(method = Image.FLIP_TOP_BOTTOM)
display(flippedVertically)
```

```python
flippedHorizontally = image.transpose(method = Image.FLIP_LEFT_RIGHT)
display(flippedHorizontally)
```

```python
rotatedLeft90 = image.transpose(method = Image.ROTATE_90)
display(rotatedLeft90)
```

```python
rotatedRight90 = image.transpose(method = Image.ROTATE_270)
display(rotatedRight90)
```

```python
targetWidth = image.width // 2
targetHeight = image.height // 2
resizedHalf = image.resize((targetWidth, targetHeight))
display(resizedHalf)
```

```python
image.close()
flippedVertically.close()
flippedHorizontally.close()
rotatedLeft90.close()
rotatedRight90.close()
resizedHalf.close()
```
